```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data['category'].nunique()
```

```python
data['category'].value_counts()
```

```python
data['day'] = data['Date_N'].dt.day
```

```python
data.head(2)
```

```python
# Pivot Table for Category vs Revenue
pivot_table = data.pivot_table(index='day', columns='category', values='quantity', aggfunc='sum')
pivot_table
```

```python
sns.heatmap(pivot_table, cmap = 'YlGnBu')
plt.title('Heatmap of Category vs Day')
plt.show()
```

> The category accounting for maximum quantity sale are fruits and vegetables, both of which are highly perishable items.

```python
# Pivot table for category vs Total
pivot_table_2 = data.pivot_table(index='day', columns='category', values='total', aggfunc='sum')
pivot_table_2
```

```python
sns.heatmap(pivot_table_2, cmap = 'RdBu')
plt.title('Heatmap of Category vs Total')
plt.show()
```

> Though the most selling categories are fruits and vegetables, the categories generating maximun revenue are kitchen, meat and seafood products. Two of which are highly perishable items.

```python
category_weekly_total = data.groupby(['category'])['total'].sum()
```

```python
category_weekly_total
```

```python
# Create bar plot
plt.bar(category_weekly_total.index, category_weekly_total.values)
plt.xlabel('Category')
plt.ylabel('Total Weekly Sales')
plt.title('Total Weekly Sales by Category')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

> The top six category by weekly sales value are -

    kitchen, meat, seafood, medicine, beverages and cleaning products
